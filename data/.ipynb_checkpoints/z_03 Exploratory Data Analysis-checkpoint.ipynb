```python
def binary_vs_target(firstFeatureName, lastFeatureName, targetName, df):
    firstIndex = df.columns.get_loc(firstFeatureName)
    lastIndex = df.columns.get_loc(lastFeatureName)
    targetIndex = df.columns.get_loc(targetName)
    
    num_plots = lastIndex - firstIndex + 1
    fig, axes = plt.subplots(num_plots // 4 + 1, 4, figsize = (10, 0.5 * num_plots))
    
    p_list = []
    
    temp_df = df
    for axesIndex, index in zip(range(0, num_plots), range(firstIndex, lastIndex + 1)):
        n1 = sum(temp_df.iloc[:,index] == 0)
        n2 = sum(temp_df.iloc[:,index] == 1)
        x1 = sum((temp_df.iloc[:,index] == 0) & (temp_df.iloc[:,targetIndex] == 0))
        x2 = sum((temp_df.iloc[:,index] == 1) & (temp_df.iloc[:,targetIndex] == 0))
        p1 = x1 / n1
        p2 = x2 / n2
        plot = sns.barplot([0,1],[p1,p2], ax = axes[axesIndex // 4][axesIndex % 4])
        if axesIndex % 4 == 0:
            plot.set_ylabel('Proportion NV\nper response')
        
        z = np.abs((p1 - p2) / np.sqrt(p1 * (1-p1) / n1 + p2 * (1-p2) / n2))
        p_value = 2 * (1 - norm.cdf(z))
        title = df.columns[index] + '\np = ' + str(p_value)
        plot.set_title(title)
        
        p_list.append([abs(p1-p2), p_value, df.columns[index]])
    fig.tight_layout()
    return p_list
```

Let's build a triangular correlation matrix to examine how all the numerical features relate to one another. The two types of numerical features are "Thermometers" and "Indices". In a "Thermometer" question, the respondent is asked to rate their favorability of a particular individual, group, or institution on a scale from 1 to 100. This scale was later reduced to increments of 5 as the vast majority of responses were given as multiples of 5. Features labeled "Index" were built from "Thermometer" data and can be thought of as summaries of particular subsets of "Thermometer" features. Thus, "Index" features will naturally have a strong correlation to the features from which they were built.

```python
firstIndex = df.columns.get_loc('VCF0204')
lastIndex = df.columns.get_loc('VCF9057')

numerical_df = df.iloc[:,range(firstIndex,lastIndex+1,2)]
numerical_df = pd.concat([numerical_df, df.iloc[:,df.columns.get_loc('VCF0649'):df.columns.get_loc('VCF1016')]])
numerical_df = numerical_df.drop(['VCF0656_dk','VCF0801_dk','VCF0649','VCF0656','VCF0723','VCF0733','VCF0801'], axis = 1)
numerical_df = numerical_df.drop(['VCF1015','VCF0138','VCF0291'], axis = 1)

numerical_df.columns = ['Thermometer: Catholics','Thermometer: Blacks','Thermometer: Whites','Thermometer: Big Business',
                       'Thermometer: Labor Unions','Thermometer: Liberals','Thermometer: Conservatives',
                        'Thermometer: Military','Thermometer: Hispanics','Thermometer: Democratic Party',
                        'Thermometer: Middle Class','Thermometer: People on Welfare','Thermometer: Poor People',
                       'Thermometer: Republican Party','Thermometer: Asian Americans','Thermometer: Congress',
                       'Thermometer: Federal Government','Thermometer: LGBT','Thermometer: Illegal Aliens',
                       'Thermometer: Christian Fundamentalists','Thermometer: Feminists','Thermometer: Both Parties',
                       'Thermometer: Both Presidential Candidates','Index: Both Presidential Candidates',
                        'Thermometer: Both VP Candidates','Index: Both VP Candidates',
                       'Thermometer: Democratic Presidential Candidate','Thermometer: Democratic VP Candidate',
                       'Thermometer: Republican Presidential Candidate','Thermometer: Republican VP Candidate',
                       'Thermometer: US President','Thermometer: US Vice President','Thermometer: Hillary Clinton',
                        'Thermometer: Democratic House Candidate', 'Thermometer: Republican House Candidate',
                        'Thermometer: Incumbent House Candidate', 'Thermometer: Challenger House Candidate',
                       'Thermometer: Supreme Court','Thermometer: Democratic Senate Candidate',
                       'Thermometer: Republican Senate Candidate']
                        
fig, axes = plt.subplots(1,1,figsize = (12,10))
numerical_df = numerical_df.loc[:,np.logical_not(numerical_df.columns.str.contains('Index'))]
corr = numerical_df.corr()
mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, ax = axes, mask = mask)
```

## Most correlations are as expected:
* There is a moderately strong positive correlation between favorability towards a racial minority and that of other minorities
* There is a moderate positive correlation between favorability towards big business and that of conservatives
* There is a moderate positive correlation between favorability towards labor unions and that of liberals
* There is a strong positive correlation between favorability towards a party and that of the same party's candidates. This correlation is strongest for presidential and vice presidential candidates, weaker for Senate candidates, and weakest for House candidates.
* There is a strong negative correlation between favorability towards a party and that of the opposite party's candidates. This correlation is strongest for presidential and vice presidential candidates, weaker for Senate candidates, and weakest for House candidates.
* There is a strong positive correlation between favorability towards a party's presidential candidate and that of the same party's vice presidential candidate.

Perhaps most interesting is the fact that while correlations between features measuring partisan support are unsurprisingly strong, correlations between features measuring idealogy and opinions of demographic groups are rather weak. This supports many political scientist's notions that politics is far more about partisan identity rather than ideology or issues.

While the above correlation matrix may not shed much new insight into the state of American politics, it does suggest that there exist some number of redundant features in the dataset.

### Thermometer Intensity and Propensity to Vote
Typically, people with more intense opinions might be more motivated to vote compared to those who are indifferent. Because Thermometer features are measured such that 50 represents a neutral opinion, the intensity of sentiment (without consideration for whether it is positive or negative) can be measured by the distance from 50. This distance was then divided by 10 for visual simplicity, resulting in intensity values from **0 to 5**.

```python
firstIndex = df.columns.get_loc('VCF0204')
lastIndex = df.columns.get_loc('VCF9057')
targetIndex = df.columns.get_loc('VCF0702')

temp_df = df.iloc[:,range(firstIndex,lastIndex+1,2)]

firstIndex = temp_df.columns.get_loc('VCF0204')
lastIndex = temp_df.columns.get_loc('VCF9057')

num_plots = lastIndex - firstIndex + 1
fig, axes = plt.subplots(num_plots, 2, figsize = (10, 4 * num_plots))

temp_df.columns = ['Thermometer: Catholics','Thermometer: Blacks','Thermometer: Whites','Thermometer: Big Business',
                       'Thermometer: Labor Unions','Thermometer: Liberals','Thermometer: Conservatives',
                        'Thermometer: Military','Thermometer: Hispanics','Thermometer: Democratic Party',
                        'Thermometer: Middle Class','Thermometer: People on Welfare','Thermometer: Poor People',
                       'Thermometer: Republican Party','Thermometer: Asian Americans','Thermometer: Congress',
                       'Thermometer: Federal Government','Thermometer: LGBT','Thermometer: Illegal Aliens',
                       'Thermometer: Christian Fundamentalists','Thermometer: Feminists','Thermometer: Both Parties',
                   'Thermometer: Major Party Index',
                       'Thermometer: Both Presidential Candidates','Index: Both Presidential Candidates',
                        'Thermometer: Both VP Candidates','Index: Both VP Candidates',
                       'Thermometer: Democratic Presidential Candidate','Thermometer: Democratic VP Candidate',
                       'Thermometer: Republican Presidential Candidate','Thermometer: Republican VP Candidate',
                       'Thermometer: US President','Thermometer: US Vice President','Thermometer: Hillary Clinton',
                        'Thermometer: Democratic House Candidate', 'Thermometer: Republican House Candidate',
                        'Thermometer: Incumbent House Candidate', 'Thermometer: Challenger House Candidate',
                       'Thermometer: Supreme Court','Thermometer: Democratic Senate Candidate',
                       'Thermometer: Republican Senate Candidate']

target0_df = temp_df[df.iloc[:,targetIndex] == 0]
target1_df = temp_df[df.iloc[:,targetIndex] == 1]

p_list = []

for axesIndex, index in zip(range(0, num_plots), range(firstIndex, lastIndex + 1)):

    g_0 = (abs(target0_df.iloc[:,index] - 50) // 10).value_counts().sort_index().plot(kind = 'bar',
            ax = axes[axesIndex][0])
    g_1 = (abs(target1_df.iloc[:,index] - 50) // 10).value_counts().sort_index().plot(kind = 'bar',
            ax = axes[axesIndex][1])
    
    title_0 = 'Did note vote\n' + temp_df.columns[index]
    g_0.set_title(title_0)
    axes[axesIndex][0].set_xlabel('Sentiment intensity')
    axes[axesIndex][0].set_ylabel('Number of respondents')
    
    title_1 = 'Did vote\n' + temp_df.columns[index]
    g_1.set_title(title_1)
    axes[axesIndex][1].set_xlabel('Sentiment intensity')
    axes[axesIndex][1].set_ylabel('Number of respondents')
    
    data1 = abs(target0_df.iloc[:,index] - 50) // 10
    data2 = abs(target1_df.iloc[:,index] - 50) // 10
    mu1 = data1.mean()
    mu2 = data2.mean()
    n1 = len(data1)
    n2 = len(data2)
    v1 = np.std(data1) ** 2
    v2 = np.std(data2) ** 2
    d = mu1-mu2
    
    t = d / np.sqrt(v1/n1 + v2/n2)
    p = (1 - stats.t.cdf(abs(t), min(n1,n2)-1)) * 2
    p_list.append([d,p,temp_df.columns[index]])
    
plt.tight_layout()
```

Rather notably, there are many '_dk' features for which almost half of all respondents who answered 'Don't know' ultimately did not vote in the election. These features all test for general knowledge about the major political parties. Interestingly, '_dk' features that test for general knowledge about the major party candidates are less predictive.
* VCF0503 = Scale, liberal-conservative: Democratic Party
* VCF0504 = Scale, liberal-conservative: Republican Party
* VCF0513 = Scale, guaranteed job-get ahead by self: Democratic Party

* VCF0541 = Scale, gvmt services: Democratic Party
* VCF0542 = Scale, gvmt services: Republican Party
* VCF0549 = Scale, defense spending: Democratic Party
* VCF0550 = Scale, defense spending: Republican Party

It is also notable that for these political party general knowledge questions, those who responded with neutral choices or choices that run counter to conventional wisdom were more likely to have not voted. For example, VCF0503 asks respondents to indicate how liberal or conservative they believe the Democratic Party to be. Almost one-third of the respondents that answered "Moderate" were non-voters while those that answered some form of "Conservative" were more likely to have not voted than those that answered some form of "Liberal." This pattern holds true for questions about the positions of major party candidates as well **(VCF9081-VCF9095)**.

#### VCF0886 - VCF9049
In general, respondents who favor increasing federal spending on various projects (with the exception of science/technology) were more likely to stay at home on election day.

#### VCF0110 - VCF0148
It appears that level of education, church attendance, and social class are all strongly negatively correlated with non-voting behavior. In other words, respondents with lower levels of education, lower frequency of church attendance, and lower social class, were far more likely to not vote.

#### VCF0301 - VCF0310
Unsurprisingly, respondents with low strength of partisanship and low interest in the elections were far more likely to not vote.
